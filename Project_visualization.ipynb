## DATA VISUALIZATION

### 1.CLOSING PRICE OVER THE TIME

```python
plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Close'], color='blue', label="Close Price")
plt.title("Tata Motors Closing Price (1995â€“2025)", fontsize=14, weight="bold")
plt.xlabel("Year")
plt.ylabel("Price (INR)")
plt.legend()
plt.show()
```

### INSIGHT : SHOWS LONGTERM GROWTH ,BIG DROPS IN 2008 AND 2020

## 2. OPEN V/S CLOSE PRICES

```python
avg_open = df['Open'].mean()
avg_close = df['Close'].mean()
plt.figure(figsize=(6,6))
plt.pie([avg_open, avg_close], labels=['Average Open', 'Average Close'],
        autopct='%1.1f%%', colors=['orange','blue'], startangle=90, explode=[0.05,0])
plt.title("Average Open vs Close Price Contribution", fontsize=14, weight="bold")
plt.show()
```

### INSIGHT : Shows whether Tata Motors tends to close higher/lower relative to its opening price on average,Gaps shows the voitle session

## HIGH V/S LOW PRICES

```python
plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['High'], label="High Price", color="green")
plt.plot(df['Date'], df['Low'], label="Low Price", color="red")
plt.title("Tata Motors High vs Low Prices", fontsize=14, weight="bold")
plt.xlabel("Year")
plt.ylabel("Price (INR)")
plt.legend()
plt.show()
```

### INSIGHT: Captures daily trading ranges,if highs dominate, stock had strong rallies; if lows dominate, more down pressure.

## TRADING VALUE OVERTIME

```python
plt.figure(figsize=(14,6))
plt.bar(df['Date'], df['Volume'], color="purple", alpha=0.6)
plt.title("Tata Motors Trading Volume (1995â€“2025)", fontsize=14, weight="bold")
plt.xlabel("Year")
plt.ylabel("Volume")
plt.show()
```

### INSIGHT : Spikes during crises â†’ panic/institutional trading.

## CORRELATION

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['Open','High','Low','Close','Volume']].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Stock Features", fontsize=14, weight="bold")
plt.show()
```

### INSIGHT : Prices (Open/High/Low/Close) are strongly correlated, Volume independent.

## YEARLY AVERAGE CLOSING PRICE


```python
yearly_avg = df.groupby('Date')['Close'].mean()
plt.figure(figsize=(12,6))
sns.lineplot(x=yearly_avg.index, y=yearly_avg.values, marker="o", color="teal")
plt.title("Yearly Average Close Price of Tata Motors", fontsize=14, weight="bold")
plt.xlabel("Year")
plt.ylabel("Average Close Price (INR)")
plt.show()
```

### INSIGHT : Clear bullish recovery after 2020.

## CLOSING PRICES BY THE YEAR


```python
plt.figure(figsize=(14,6))
sns.boxplot(x="Date", y="Close", data=df, palette="Set3")
plt.xticks(rotation=90)
plt.title("Distribution of Tata Motors Closing Prices by Year", fontsize=14, weight="bold")
plt.show()
```

### INSIGHT:Some years had wide fluctuations (2008, 2020).

## CLOSE V/S VOLUME


```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Volume'], y=df['Close'], alpha=0.5, color="darkblue")
plt.title("Scatterplot: Close Price vs Trading Volume", fontsize=14, weight="bold")
plt.xlabel("Volume")
plt.ylabel("Close Price (INR)")
plt.show()
```

## AVERAGE CLOSING PRICES MONTHLY

### INSIGHT:useful for traders to identify strong/weak months.
